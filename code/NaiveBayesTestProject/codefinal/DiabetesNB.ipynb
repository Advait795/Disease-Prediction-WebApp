Load Data

```python
import pandas as pd

data = pd.read_csv(r"C:\Users\ADWAIT\Desktop\Final Year Project\Data\disease Prediction\Hypertension Stroke Diabetes\diabetes_data.csv")

data.head(15)
```

Handling Missing Values

```python
data.fillna(data.median(), inplace=True)
```

Encoding Categorical Variables

```python
categorical_columns = ['Sex', 'HighChol', 'CholCheck', 'Smoker', 'HeartDiseaseorAttack', 'PhysActivity', 
                       'Fruits', 'Veggies', 'HvyAlcoholConsump', 'DiffWalk', 'Stroke', 'HighBP' ]

for col in categorical_columns:
    data[col] = data[col].astype('category')

data.head(10)
```

Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

num_cols = ['Age', 'BMI', 'GenHlth', 'MentHlth', 'PhysHlth']
scale = MinMaxScaler()

data[num_cols] = scale.fit_transform(data[num_cols])

data.head(10)
```

Splitting Data


```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, chi2

X = data.drop(columns=['Diabetes'])
y= data['Diabetes']

selector = SelectKBest(score_func=chi2, k=16)
X_new = selector.fit_transform(X, y)

selected_fea = X.columns[selector.get_support(indices=True)]
print(selected_fea)


X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.15, random_state=7)

print(X_new[:10])
```

Addressing Class Imbalance

```python
from imblearn.over_sampling import SMOTE


smote = SMOTE(random_state=7)
X_train_re, y_train_re = smote.fit_resample(X_train, y_train)


# categorical_col = ['HighChol', 'HeartDiseaseorAttack', 'GenHlth', 'DiffWalk', 'HighBP' ]

# cat_imputer = SimpleImputer(strategy='most_frequent')
# # X_train_re[categorical_col] = cat_imputer.fit_transform(X_train_re[categorical_col])


# has_missing_val = X_train_re.columns[X_train_re.isna().any()]

# print(has_missing_val)

```

Model Training & Evaluation



```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

model = GaussianNB()

model.fit(X_train_re, y_train_re)

y_predict = model.predict(X_test)

print("Accuracy: ", accuracy_score(y_test, y_predict))
print("confusion Matrix: \n \n", confusion_matrix(y_test, y_predict))
print("Classification Report: \n\n", classification_report(y_test, y_predict))

```
